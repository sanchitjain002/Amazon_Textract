```python
import boto3
import json
import sys
from pprint import pprint
#from trp import Document
import sys
from PIL import Image, ImageDraw
import csv
```

```python
textract = boto3.client(
   service_name = 'textract',
    region_name = 'us-east-2'
)

s3 = boto3.resource('s3')
```

```python
client = boto3.client('s3')
paginator = client.get_paginator('list_objects_v2')
result = paginator.paginate(Bucket='sanchit002',StartAfter='RC/')
All = []
for page in result:
    if "Contents" in page:
        for key in page[ "Contents" ]:
            keyString = key[ "Key" ]
            All.append(keyString)
          #  print (keyString)
```

```python
for i in range(len(All)):
    response = textract.analyze_document(
       Document = {
           'S3Object' : {
               'Bucket' : 'sanchit002',
               'Name' : All[i] 
           }
       },
         FeatureTypes=["FORMS"] )
    #print(response)
    Filename = All[i]
    with open('output.csv', mode='a', newline='') as aws_field_file:
        field_write = csv.writer(aws_field_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        field_write.writerow([Filename])
        field_write.writerow(" ")
        field_write.writerow(" ")
        print(Filename)
        print(" ")
        print(" ")
        for item in response["Blocks"]:
            if item["BlockType"] == "LINE" :
                field_write.writerow([item["Text"]])
                print ('\033[94m' +  item["Text"] + '\033[0m')
        print("--------------------------------------")
        print(" ")
        print(" ")
        field_write.writerow(" ")
        field_write.writerow(["-------------------------------------------------------"])
        field_write.writerow(" ")

        
#To find the key, value pair

#doc = Document(response)
#with open('output.csv', mode='w') as aws_field_file:
#    field_write = csv.writer(aws_field_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
 
#    for page in doc.pages:
#        print("Fields:")
#        for field in page.form.fields:
#            field_write.writerow([field.key, field.value])     
#            print("Key: {}, Value: {}".format(field.key, field.value))
3
```
